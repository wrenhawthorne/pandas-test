```python
# filter to just the local auth
df[df["Proprietorship Category (1)"] == "Local Authority"]
```

```python
# a single & is an AND operator
df[(df["Proprietorship Category (1)"] == "Limited Company or Public Limited Company") & (df["Proprietorship Category (1)"] != "Local Authority")]
```

```python
# a single pipe | is an OR operator
df[(df["Proprietorship Category (1)"] == "Limited Company or Public Limited Company") | (df["Proprietorship Category (1)"] != "Local Authority")]
```

```python
df[df["Proprietor Name (1)"].str.startswith("A")]
```

```python
# more complex filter
df[df["Proprietorship Category (1)"].isin(["Corporate Body", "Limited Company or Public Limited Company", "Community Benefit Society (Company)", "Registered Society (Corporate Body)"])]
```

```python
# tilde ~ makes it 'is not in'
df[~df["Proprietorship Category (1)"].isin(["Corporate Body", "Limited Company or Public Limited Company", "Community Benefit Society (Company)", "Registered Society (Corporate Body)"])]
```
